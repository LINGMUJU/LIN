```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取Excel文件
excel_file = 'pydata02.xlsx'  # 替换为你的Excel文件路径
df = pd.read_excel(excel_file)

# 设置日期列为索引（如果尚未是索引）
df.set_index('日期', inplace=True)

# 设置滑动平均窗口大小和权重参数
window_size = 3  # 滑动平均窗口大小
alpha = 0.05  # 权重参数，可以根据需要调整

# 计算一阶滑动平均
df['滑动平均'] = df['大气温度'].rolling(window=window_size).apply(lambda x: x[-1] * alpha + x[:-1].mean() * (1 - alpha), raw=True)

# 绘制原始数据和滑动平均曲线
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['大气温度'], label='原始数据')
plt.plot(df.index, df['滑动平均'], label='滑动平均')
plt.xlabel('日期')
plt.ylabel('大气温度')
plt.legend()
plt.title('一阶滑动平均预测')
plt.grid(True)

# 显示图表
plt.show()
```
